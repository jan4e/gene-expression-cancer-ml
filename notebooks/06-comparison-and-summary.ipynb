## Reflection

**Which model performed best?**  
The Neural Network (ANN) was the best with **100% accuracy and F1-score** — no mistakes at all on the test set.  
Logistic Regression and SVM were almost as good (only 1 error each).

**Which was easiest?**  
Logistic Regression was the easiest and fastest to train — it converged quickly and needed almost no tuning.  
SVM took longer to train (because of high dimensions), and ANN needed more setup (label encoding, epochs, etc.).

**Did ANN really help?**  
Yes, but only a little — it reached perfect score, while others were already at 99.38%.  
The dataset is very separable (gene patterns clearly distinguish the 5 cancer types), so even simple models work great.  
ANN didn't give a huge improvement here, but it was worth adding as requested by the professor.

**Overall thoughts** (from proposal):  
I expected baseline to be reasonable but limited, and advanced models better — especially after feature reduction.  
In reality, all models were excellent even without PCA.  
Next experiments should include PCA to see if we can keep ~99–100% with much fewer genes (dimensionality reduction).

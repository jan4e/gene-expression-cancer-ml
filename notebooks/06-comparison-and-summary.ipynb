## Reflection

**Which model performed best?**  
Logistic Regression and SVM achieved the highest test performance (~99.4% accuracy), making them the best-performing models on this dataset.

The Neural Network (ANN) also performed very well (~97% accuracy), but slightly worse than the linear models, likely due to mild overfitting given the extremely high number of features compared to samples.


**Which was easiest?**  
Logistic Regression was the easiest and fastest to train, it converged quickly and needed almost no tuning.  
SVM took longer to train (because of high dimensions), and ANN needed more setup (label encoding, epochs, etc.).

**Did ANN really help?**  
Not significantly. The ANN achieved strong performance (~97% accuracy), but it did not outperform
Logistic Regression and SVM, which already achieved ~99.38% accuracy.  

The dataset is very separable (gene expression patterns clearly distinguish the five cancer types),
so even simple linear models perform extremely well.  

Although the ANN did not improve test performance compared to simpler models, it still demonstrated
the ability to learn complex patterns and fulfilled the requirement to apply artificial neural
networks to biological data.


**Overall thoughts**:  
I expected baseline to be reasonable but limited, and advanced models better, especially after feature reduction.  
In reality, all models were excellent even without PCA.
